```python
# it is importing the library called pandas , to work with structured data like tables, spreadsheets and SQL Data
# and we are naming it as 'pd' for ease of calling
import pandas as pd
```

```python
# declaring a variable called dataset and reading the file using pandas to analyse/process the data 
dataset = pd.read_csv("Salary_Data.csv")
```

```python
# it will print the whole Salary Data as it present in the excel 
dataset
```

```python
# it will help you get only the input we are going to use in the project
independent = dataset[["YearsExperience"]]
```

```python
# to print the input needed for the project
independent
```

```python
# this is to print the output or the number that we want to predict using this data
dependent  = dataset[["Salary"]]
dependent
```

```python
# Splitting train and test set
# sklearn is for importing machine learning pre data set (functions)
#it is helpful to split the train and test sets using built in functions
from sklearn.model_selection import train_test_split
```

```python
# now we have splitted data and stored on four different variables
# meaning we have splitted independent and dependent variable by 70% and 30% each,to train a model and test a model
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
# Using the sklearn library import the linear regression file or inbuilt class
# useful for accessing linear based models
from sklearn.linear_model import LinearRegression
```

```python
# it is a linear regression inbuilt class used to store all the data's of the linear model - object name = regressor
# the class can now be able to use under the object name regressor()
regressor = LinearRegression()
```

```python
# now using a inbuilt function ".fit" we are using the training data to fit inside the model 
# it fits perfectly it will display -> LinearRegression
regressor.fit(X_train,Y_train)
```

```python
#now we are calculating weight(constant difference between two points) , coef_ is inbuilt name for weight
weight  = regressor.coef_
weight
```

```python
# it is used to calculate the initial value
bais = regressor.intercept_
bais
```

```python
# we are now predicting the output of y by giving the X_test, which is basically 70% percent of the data
# after this , if the Y_pred produces the results if which is closer to the 30 percent of the value it is a good model else not a good model
Y_pred  = regressor.predict(X_test)
```

```python
# now we are importing one of the evaluating metrices to check whether the built model is good or not
# for deployment
from sklearn.metrics import r2_score
r_score = r2_score(Y_test,Y_pred)
```

```python
r_score
```

```python
#Helps you save and load ML models, scalers, or any Python object.
import pickle
```

```python
#creating a blank file named "..sav "  to save a regressor mmodel into this file
filename ="finalized_model_linear.sav"
```

```python
# now dumping or loading all the data into the blank file
pickle.dump(regressor,open(filename,'wb'))
```

```python
#now we are creating a varaible to load all the files and read it , so that we can acess it for models
loaded_model = pickle.load(open("finalized_model_linear.sav",'rb'))
```
